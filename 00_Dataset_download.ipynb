## Downloading images

In this Notebook we collect a series of images from Google which we are going to use in our tutorial in the next steps. For downloading the images we use a python tool call 'google-images-download'.

````
!apt-get install tree
!pip install google-images-download
````

````
!mkdir dataset
````

As a next step, we create a JSON file which serves as an inpute for the google image downloader. We provide the search term as well as the amount of images we want to fetch. 

````
%%writefile dataset/download_config.json

{
    "Records": [
        {
            "keywords": "Tie Dye sneakers",
            "limit": 50,
            "print_urls": true
        },
        {
            "keywords": "converse chuck 70",
            "limit": 50,
            "print_urls": true
        },
        {
            "keywords": "Exaggerated Soles",
            "limit": 50,
            "print_urls": true
        },
        {
            "keywords": "nike air max 90 essentials",
            "limit": 50,
            "print_urls": true
        },
        {
            "keywords": "nike react",
            "limit": 50,
            "print_urls": true
        }
    ]
}
````

````
!cat dataset/download_config.json
````

````
!googleimagesdownload -cf dataset/download_config.json -o dataset
````
